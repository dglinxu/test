```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.sans-serif']='microsoft yahei'
```

```python
x=np.linspace(0,np.pi*2)
plt.figure(figsize=(8,6))
y=np.sin(x)
y2=np.cos(x)
plt.grid(True,axis='both')
plt.plot(x,y,x,y2)
plt.text(0,-1,s='文字说明',fontsize=20,color='r')
```

```python
plt.figure(figsize=(16,3))
ax=plt.subplot(1,3,1)
ax.plot(x,y,x,y2)
ax.grid(color='r')

ax=plt.subplot(1,3,2)
ax.plot(x,y,x,y2)
ax.grid(color='g')

ax=plt.subplot(1,3,3)
ax.plot(x,y,x,y2)
ax.grid(color='y',linestyle='-.')
```

```python
plt.plot(x,y,x,y2)
plt.axis([0,6,-1,1])
plt.grid(color='g',linestyle='-')
plt.xlabel('---X轴---',fontsize=20)
plt.ylabel('------Y轴------',fontsize=20,rotation=90,color='y')
plt.title('正弦、余弦曲线',fontsize=25,color='pink')
plt.legend(['正弦','余弦'],loc='upper right')
```
