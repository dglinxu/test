```python
def capitalize(s):
    pass
    n=len(s)
    even=''
    odd=''
    for i in range(n):
        if i%2==0:
            even+=s[i].capitalize()
            odd+=s[i]
        else:
            even+=s[i]
            odd+=s[i].capitalize()
    return [even,odd]
capitalize("abcdef")
```

```python
def pig_it(text):
    #your code here
    words=text.split()
    res=[w[1:]+w[0]+'ig' if ord(w[0])>64 else w for w in words]
    return ' '.join(res)
pig_it('This is my string !')
```

```python
def generate_hashtag(s):
    #your code here
    if s=='':
        return False
    words=s.split()
    w=[w.capitalize() for w in words if w.isalpha()]
    res='#'+''.join(w)
    if len(res)>140:
        return False
    return res
generate_hashtag('Do We have A @ Hashtag')
```

```python
def rot13(message):
    sl='abcdefghijklmnopqrstuvwxyz'
    su=sl.upper()
    s=''
    for c in message:
        if c in su:
            n=(su.index(c)+13)%26
            s+=su(n)
        elif c in sl:
            n=(sl.index(c)+13)%26
            s+=sl(n)
        else:
            s+=c
    return s
rot13("test")
```

```python
def changN(n,m,num):
    s='0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    ans=''
    k=0
    for i in str(num):
        k=k*n+int(i)
    while k>=m:
        ans=s[k%m]+ans
        k=k//m
    return s[k]+ans
n,m=tuple(map(int,input().split()))
num=int(input())

print(changN(n,m,num))
    
```
